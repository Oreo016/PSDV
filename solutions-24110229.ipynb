```python
l=[]
for i in range(1,21):
    l.append(i)
sum_even=0
sum_odd=0
for i in l:
    if i%2==0:
        sum_even+=1
    else:
        sum_odd+=1
mean=(sum_even+sum_odd)/2
if len(l)%2!=0:
    median=l[int((len(l)+1)/2)]
else:
    median=(l[int(len(l)/2)]+l[int(len(l)/2+1)])/2
s_d=0
for i in l:
    s_d+=abs(mean-i)
s_d=s_d/len(l)
print("mean",mean)
print("median",median)
print("standard deviation",s_d)
```

```python
def anagram_finder(a,l):
   a_=[]
   final=[]
   for i in a:
      if i!=" ":
         a_.append(i)
   for i in l:
      ch=[]
      
      for j in i:
         if j!=" ":
            ch.append(j)
      if sorted(ch)==sorted(a_):
        
         final.append(i)
      
    
   return final
print(anagram_finder("listen",["enlist", "google", "inlets", "banana"]))
```

```python
import matplotlib.pyplot as pl
def normalise(l):
    l.sort()
    
    k=[]
    for i in l:
        k.append((i-l[0])/(l[len(l)-1]-l[0]))
    pl.plot(l,k)
    pl.show()
    return k
print(normalise([2,4,6]))
```
